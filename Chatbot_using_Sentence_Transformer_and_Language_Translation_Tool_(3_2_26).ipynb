```python
df=pd.DataFrame(faq_data)
df
```

```python
from sentence_transformers import SentenceTransformer
model=SentenceTransformer("distiluse-base-multilingual-cased-v1")
faq_emb=model.encode(faq_data['question'])
```

```python
from sentence_transformers import util
query=input("Ask a question:")
query_emb=model.encode(query)
closest=util.semantic_search(query_emb,faq_emb,top_k=1)
idx=closest[0][0]['corpus_id']
df["answer"][idx]
```

```python
def dynamic_faq_system():
  print("Chatbot:Hello ! Ask a question or type 'exit' to quit.")
  while True:
    user_query=input("You:")
    if user_query.lower() in ["exit","quit"]:
      print("Chatbot:Goodbye!")
      break
      #Encode user_query.
    query_emb=model.encode(user_query)
    #Finally find the closest Answer of given prompt.
    closest=util.semantic_search(query_emb,faq_emb,top_k=1)
    best_match_id=closest[0][0]['corpus_id']
    best_answer=df.iloc[best_match_id]['answer']
    #Display the answer
    print("Chatbot:",best_answer)
```

```python
dynamic_faq_system()
```

```python
from deep_translator import GoogleTranslator as gt
```

```python
text= "I am a nice Human being"
```

```python
gt(source = "english",target="hindi").translate(text)
```

```python
def trans(text,target):
  trans_text=gt(target=target).translate(text)
  return trans_text
```

```python
while True:
  print("\n Language Translation Tool")
  input_text=input("Enter your text" )
  if input_text.lower()=="quit":
    break
  target_lang=input("To translate lang,(en,hi,ur,ta,punjabi,)")
  trans_text=trans(input_text,target_lang)
  print("Result : ",trans_text)
```
